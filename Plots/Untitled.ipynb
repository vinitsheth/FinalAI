```python
import pickle
```

```python
import matplotlib.pyplot as plt
```

```python
path1 = 'lifelonglpaExpanded.pickle'
fil1 = open(path1,'rb')
lifelonglpaExpanded = pickle.load(fil1)
fil1.close()
```

```python
path1 = 'lifelongdstarExpanded.pickle'
fil1 = open(path1,'rb')
lifelongdstarExpanded = pickle.load(fil1)
fil1.close()
```

```python
plt.plot(range(1,len(lifelonglpaExpanded)+1),lifelonglpaExpanded,label='LPA*')
plt.plot(range(1,len(lifelongdstarExpanded)+1),lifelongdstarExpanded,label='d*')
plt.title('Maze size 19 X 16')
plt.ylabel('Number Of Expanded Nodes')
plt.xlabel('Walls')
plt.legend()
plt.show()
```
